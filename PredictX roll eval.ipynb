```python
def haversine_np(Departure_lon, Departure_lat, Arrival_lon, Arrival_lat):
    """
    Calculate the great circle distance between two points
    on the earth (specified in decimal degrees)

    All args must be of equal length.    

    """
    Departure_lon, Departure_lat, Arrival_lon, Arrival_lat = map(np.radians, [Departure_lon, Departure_lat, Arrival_lon, Arrival_lat])

    dlon = Arrival_lon - Departure_lon
    dlat = Arrival_lat - Departure_lat

    a = np.sin(dlat/2.0)**2 + np.cos(Departure_lat) * np.cos(Arrival_lat) * np.sin(dlon/2.0)**2

    c = 2 * np.arcsin(np.sqrt(a))
    km = 6367 * c
    return km

df['distance'] = haversine_np(df['Departure_lon'],df['Departure_lat'],df['Arrival_lon'],df['Arrival_lat'])
```
